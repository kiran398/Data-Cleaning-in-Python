```python
# gives total number of duplicated rows 
df.duplicated().sum()
```

```python
# to drop or dlt the duplicated values
df = df.drop_duplicates()
df
```

```python
# drop some specific columns fromt the data set and here axis=1 represents the columns
df =  df.drop(['Not_Useful_Column'], axis = 1)
```

```python
# there were some unwanted characters in the last_name col that need to be deleted, so we use strip that help to dlt the white space but in this case 
#we are specifying what needs to be deleted
df["Last_Name"] = df["Last_Name"].str.strip('...')
df["Last_Name"] = df["Last_Name"].str.strip('/')
df["Last_Name"] = df["Last_Name"].str.strip('_')
 # OR
```

```python
df
```

```python
pd.isnull(df).sum()
```

```python
# here we need to make all the numubers in one format where at first we are replacing that characters that are not in a-zA-Z0-9 into nothing. Then we are 
# change all into string, after which we are puttong - - - after each 3 numbers to make all in same format. At last we replace the Na or nan value with empty row cells

df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '_' + x[3:6] + '_' + x[6:10])
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
df['Phone_Number']=df['Phone_Number'].str.replace('Na--','')
df

```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('__','')
```

```python
# Here we are spliting the Address and saving into new diff columns
df[["Street_Address","States","ZipCode"]] = df["Address"].str.split(',',n=2, expand=True)
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"]=df["Paying Customer"].str.replace('No','N')
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
```

```python
# this simply replaces the NaN value to empty
df = df.fillna('')
```

```python
# with the help of loop and the index we are looking if Donotcontact is yes then we are delecting the entire row of that individual
for x in df.index:
    if df.loc[x,'Do_Not_Contact'] == 'Y':
        df.drop(x,inplace=True)
```

```python
for x in df.index:
    if df.loc[x,'Phone_Number'] == '':
        df.drop(x,inplace=True)
```

```python
df = df.drop(['ZipCode'],axis=1)
```
